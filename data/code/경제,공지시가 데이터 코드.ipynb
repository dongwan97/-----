# 미래사회 문제 해결 공모전 데이터 병합


### 데이터 불러오기

```python
import pandas as pd
```

```python
gdp=pd.read_excel("./data/분기별 GDP,경제성장률(2020~2023).xlsx")
customer = pd.read_excel("./data/소비자물가지수(2020~2023).xlsx")
apartment_avg=pd.read_excel("./data/아파트 공지시가(2020~2023).xlsx")
# apartment_cat=pd.read_excel("./data/주택종류 구별(2020~2023).xlsx")
```

```python
gdp.rename(columns={'Unnamed: 0':'시점'},inplace=True)
gdp.info()
```

```python
customer.rename(columns={'Unnamed: 0':'시점'},inplace=True)
customer.info()
```

```python
apartment_avg.info()
```

```python
df=pd.merge(gdp,customer,how='outer',on='시점')
df=pd.merge(df,apartment_avg,how='outer',on='시점')
df.sort_values('시점')
```

```python
all_columns = ['시점','국내총생산(명목GDP)','경제성장률(실질GDP성장률)','소비자물가','- 농축수산물','- 공업제품',
           '- 집세','- 공공서비스','- 개인서비스','근원물가','생활물가','공지시가_전국','공지시가_수도권',
           '공지시가_지방','공지시가_서울','공지시가_도심권','공지시가_동북권','공지시가_동남권',
           '공지시가_서북권','공지시가_서남권']
columns = ['시점','국내총생산(명목GDP)','경제성장률(실질GDP성장률)','소비자물가','근원물가','생활물가',
           '공지시가_서울','공지시가_도심권','공지시가_동북권','공지시가_동남권',
           '공지시가_서북권','공지시가_서남권']
```

```python
df_final=df[columns]
df_final
```

```python
df_final.to_csv('./data/경제_공지시가 데이터.csv',encoding='cp949',index=False)
```
